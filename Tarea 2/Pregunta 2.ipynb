## 2. Redes recurrentes sobre texto

````python
import numpy as np
import pandas as pd
df_ner = pd.read_csv("ner.csv", error_bad_lines=False, engine='python')
df_ner.dropna(inplace=True)
````

<a id="pregunta_a"></a>
> a) _POS tagging_ corresponde a la tarea de identificar _a que parte del discurso (part of speech)_ pertenece cada palabra dentro de un texto, a partir de su contexto y definición. <br/>
Para realizar esta tarea es necesario usar una red recurrente ya que es relevante tener en consideración el contexto de cada palabra, es decir, conocer la secuencia que le precede. <br/>
Se extraerán 4 columnas del _dataset_, _lemma, pos, tag_ y _prev-oib_. Las cuales corresponden al lemma de la palabra, al _pos tag_ del _token_, a la notación _IOB_ de la palabra y la notación _IOB_ del _token_ anterior. <br/>
En el ejercicio se decidió realizar un _POS taggin_ en vez de un _NER_ por ello la columna _tag_ no será utilizada. <br/>
Un _lemma_ es la forma inflexionada de una palabra. Un _token_ puede contener distintas flexiones, como lo pueden ser las conjugaciones de un verbo. Sin embargo sería útil poder condensar todas estas variaciones en una sola palabra para simplificar los análisis. Por este último argumento sería mejor utilizar _lemmas_ a las palabras mismas ya que implica que la red no necesita aprender tags para cada flexión de una palabra.

````python
dataset = df_ner.loc[:,["lemma","pos","tag","prev-iob"]]
lemmas, labels = set(), set()
dataX, dataY = [], []
````

````python
print(list(dataset))
for fila in dataset.values:
    l = ""
    for i_cell, v_cell in enumerate(fila):
        l += "[{}: {}] ".format(i_cell, v_cell)
    print(l)
    break
````

````python
frecuencias = {}
sentence= []
labels_sentence = []
max_sentence_length = 0
min_sentence_length = -1
marcas_crippl = []
for fila in dataset.values:
    if fila[-1] == "__START1__": 
        if len(sentence) > max_sentence_length:
            max_sentence_length = len(sentence)
        if len(sentence) != 0:
            if min_sentence_length == -1:
                min_sentence_length = len(sentence)
            elif len(sentence) < min_sentence_length:
                print(sentence)
                min_sentence_length = len(sentence)
        dataX.append(np.asarray(sentence))
        dataY.append(np.asarray(labels_sentence))
        sentence= []
        labels_sentence = []
    if fila[0] == "crippl":
        marcas_crippl.append(len(dataX))
    lemmas.add(fila[0])
    labels.add(fila[1])
    sentence.append(fila[0])
    labels_sentence.append(fila[1])
    if fila[0] not in frecuencias.keys():
        frecuencias[fila[0]] = 1
    else:
        frecuencias[fila[0]] += 1

dataX = np.asarray(dataX[1:])
dataY = np.asarray(dataY[1:])
````

````python
print("dataX =", len(dataX))
print("max_sentence_length =", max_sentence_length)
print("min_sentence_length =", min_sentence_length)
print("lemmas =", len(lemmas))
print("labels =", len(labels))
````

<a id="pregunta_c"></a>
> c) Se añaden dos _tokens_ especiales a la lista de _lemmas_ y de _labels_, cada uno será utilizado para completar cada sentencia a un largo concreto. Dichos _tokens_ corresponden a <i>\_LEMMAPAD\_</i> y <i>\_TAGPAD\_</i>. <br/>
Es necesario transformar los _lemmas_ y _tags_ de cadenas de texto a números para que puedan ser usados con la red. PAra ello se usarán 4 diccionarios. Funcionarán como una tabla, dos de ellos (_lemma2idx_ y _lab2idx_) contienen la conversión entre la palabra y su número de identificación y los otros dos (_idx2lemma_ y _idx2lab_) contienen la conversión entre el número de identificación y la palabra. Los últimos dos son necesarios para reconstruir los resultados que entregue la red. <br>
También se realizan sobre las sentencias sus conversiones de tokens a números identificadores.

````python
lemmas.add("_LEMMAPAD_")
labels.add("_TAGPAD_")
lemmas = list(lemmas)
labels = list(labels)
lemma2idx = {l: i for i, l in enumerate(lemmas)}
lab2idx = {t: i for i, t in enumerate(labels)}
idx2lemma = {i: l for i, l in enumerate(lemmas)}
idx2lab = {i: t for i, t in enumerate(labels)}
dataX_copia = dataX.copy()
dataX = [[lemma2idx[lemma] for lemma in sentence ] for sentence in dataX]
dataY = [[lab2idx[pos] for pos in pos_tags ] for pos_tags in dataY]
n_lemmas = len(lemmas)
n_labels = len(labels)
````

<a id="#pregunta_d"></a>
> d) Ahora se realiza lo mencionado en [e)](#pregunta_c), la adición de _tokens_ a las sentencias para lograr una longitud concreta de tal forma que todas las secuencias tengan el mismo largo. <br/>
El largo de todas las secuencias se fijará en 81, valor obtenido en la pregunta [a)](#pregunta_a), que corresponde al la mayor longitud que registró una secuencia. La función ```pad_sequences``` se encarga de realizar esta tarea, tanto en el conjunto de secuencias como el en conjunto de _tags_, esto tiene sentido ya que debe existir una correspondencia 1-1 para cada elemento de los vectores. Cada _lemma_ tiene asociado un _tag_, por eso tiene sentido añadir los _tokens_ especiales mencioandos en [c)](#pregunta_c), porque aparte de dejar todos los vectores del mismo tamaño, se debe tener un _tag_ asociado a el _lemma_ especial. <br/>
Hace más sentido añadir los _tokens_ especiales en el final de las secuencias. Ya que las redes recurrentes aprenden teniendo en cuenta los valores anteriores, el poner en el inicio los valores especiales puede perjudicar el aprendizaje ya que la red los podría considerar como parte del contexto de las secuencias, no siendo este el caso. <br/>
Cabe mencionar que lo que se mencionó sobre que las redes recurrentes aprenden sobre valores pasados no se consideró a las redes bidireccionales.

````python
from keras.preprocessing.sequence import pad_sequences

X = pad_sequences(maxlen=max_sentence_length, sequences=dataX, padding="post", value=lemma2idx["_LEMMAPAD_"])
y = pad_sequences(maxlen=max_sentence_length, sequences=dataY, padding="post", value=lab2idx["_TAGPAD_"])
````

> Además se probará rellenando al inicio para probar su desempeño, para establecer una comparación con el el relleno hacia el final.

````python
X_pre = pad_sequences(maxlen=max_sentence_length, sequences=dataX, padding="pre", value=lemma2idx["_LEMMAPAD_"])
y_pre = pad_sequences(maxlen=max_sentence_length, sequences=dataY, padding="pre", value=lab2idx["_TAGPAD_"])
````

````python
for l, t in zip(X[25], y[25]):
    print("{:15}| {:15}".format(idx2lemma[l], idx2lab[t]))
````

````python
for l, t in zip(X_pre[25], y_pre[25]):
    print("{:15}| {:15}".format(idx2lemma[l], idx2lab[t]))
````

> e) A continuación se codifica el conjunto de _tags_ como vectores _one hot_, esto es necesario ya que se realizará una tarea de clasificación, de esta forma quedará un conjunto de tres dimensiones como bien es mencionado en el enunciado, una dimensión corresponde a cantidad de ejemplos, a la cantidad máxima de palabras de una secuencia y a la cantidad de _tags_. <br/> 
También se crean a partir del _dataset_ conjuntos de entrenamiento y de prueba.

````python
from keras.utils import to_categorical
y_onehot = np.asarray([to_categorical(i, num_classes=n_labels) for i in y])
y_onehot_pre = np.asarray([to_categorical(i, num_classes=n_labels) for i in y_pre])
print("Dimensiones antes de codificar como one hot {}, después de codificar como one hot {}".format(y.shape, y_onehot.shape))
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.3,random_state=22)
X_train_pre, X_test_pre, y_train_pre, y_test_pre = train_test_split(X_pre, y_onehot_pre, test_size=0.3,random_state=22)
print("Dimensiones del conjunto de secuencias {}".format(X.shape))
print("Dimensiones del conjunto de entrenamiento, secuencias: {}, tags: {}".format(X_train.shape, y_train.shape))
print("Dimensiones del conjunto de pruebas, secuencias: {}, tags: {}".format(X_test.shape, X_test.shape))
````

> f) Se define una red con una capa de _embedding_ seguida de una capa _LSTM_ de 100 neuronas para luego terminar con una capa _fully connected_ de 42 neuronas con una función de activación _softmax_ para clasificar las entradas. <br/>
El entrada de la red recibe vectores de largo 81 correspondientes a las secuencias. La capa de _embedding_ ```Embedding(input_dim=a, output_dim=b, input_length=c)``` recibe tres argumentos, el **primero** corresponde a la cantidad de palabras diferentes que hay en el _dataset_, la **segunda** indica el tamaño de los vectores que entregará la salida de la capa y el **tercero** es para indicar el tamaño que tiene cada vector de entrada.
Internamente la capa funciona como una tabla, para cada número identificatorio de una palabra en el rango del **primer** parámetro se generará un vector del largo indicado en el **segundo** parámetro. Es decir

| Vector de entrada | Vector de salida   |
|-------------------|--------------------|
| $(batch size, c)$   | $(batch size, c, b)$ |

    <tr>
        <th>Índice</th>
        <th>Embedding vector</th>
    </tr>
    <tr>
        <td>$0$</td>
        <td>$[x^{0}_{0},\ldots,x^{0}_{b-1}]$</td>
    </tr>
    <tr>
        <td>$⋮$</td>
        <td>$⋮$</td>
    </tr>
    <tr>
        <td>$a-1$</td>

````python
from keras.models import Sequential
from keras.layers import CuDNNLSTM, LSTM, Embedding, Dense, Dropout
from keras.utils.vis_utils import plot_model
````

````python
embedding_vector = 32
model_a = Sequential()
model_a.add(Embedding(input_dim=n_lemmas, output_dim=embedding_vector, input_length=max_sentence_length))
model_a.add(CuDNNLSTM(units=100,return_sequences=True))
model_a.add(Dense(n_labels, activation='softmax'))
model_a.summary()
model_a.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
plot_model(model_a, to_file='2_f.png', show_shapes=True, show_layer_names=True)

model_a_pre = Sequential()
model_a_pre.add(Embedding(input_dim=n_lemmas, output_dim=embedding_vector, input_length=max_sentence_length))
model_a_pre.add(CuDNNLSTM(units=100,return_sequences=True))
model_a_pre.add(Dense(n_labels, activation='softmax'))
model_a_pre.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
````

> g) Para comparar el impacto a la hora de usar distintos valores de embedding se ejecuta el mismo modelo de la pregunta [f)](#pregunta_f) para valores de embedding entre 8 y 256. <br/>
La función ```generar_modelo_g(embedding_vector)``` recibe como parametro un vector con distintos valores de embedding y genera modelos para todos aquellos valores. <br/>
La función ```entrenar_modelo_g(emb_model)``` recibe una tupla formada por el valor de _embedding_ que usa la red y el modelo mismo que se entrenará. <br/>
Finalmente ```graficar_g(histogramas)``` recibe una lista de las tuplas retornadas por ```entrenar_modelo_g``` para graficarlas.

````python
def generar_modelo_g(embedding_vector):
    model = Sequential()
    model.add(Embedding(input_dim=n_lemmas, output_dim=embedding_vector, input_length=max_sentence_length))
    model.add(CuDNNLSTM(units=100,return_sequences=True))
    model.add(Dense(n_labels, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return (embedding_vector, model)

def entrenar_modelo_g(emb_model):
    return (emb_model[0], emb_model[1].fit(X_train, y_train, validation_data=(X_test, y_test), epochs=12, batch_size=128))

def graficar_g(histogramas):
    fig, (ax1, ax2) = plt.subplots(2, 1)
    for ev, h in histogramas:
        ax1.plot(h.history['acc'], label='entrenamiento ' + str(ev) + " dimensiones")
        ax1.plot(h.history['val_acc'], label='validación ' + str(ev) + " dimensiones")
        
    ax1.set(ylabel='Accuracy', xlabel='Epoch')
    ax1.legend(fancybox=True, bbox_to_anchor=(1.04,1))
    for ev, h in histogramas:
        ax2.plot(h.history['loss'], label='entrenamiento ' + str(ev) + " dimensiones")
        ax2.plot(h.history['val_loss'], label='validación ' + str(ev) + " dimensiones")    
    ax2.set(ylabel='Loss', xlabel='Epoch')
    plt.show()
    
    for ev, h in histogramas:
        fig, (ax1,  ax2) = plt.subplots(1, 2, figsize=(10,3))
        ax1.plot(h.history['loss'], label='entrenamiento')
        ax1.plot(h.history['val_loss'], label='validación')
        ax1.set(ylabel='Loss', xlabel='Epoch', title='Embedding de ' + str(ev))
        ax1.grid(linestyle='-', linewidth=0.2)
        ax1.xaxis.set_ticks([v for v in range(12)])
        ax1.legend(fancybox=True, loc='upper right')
        
        ax2.plot(h.history['acc'], label='entrenamiento')
        ax2.plot(h.history['val_acc'], label='validación')
        ax2.set(ylabel='Accuracy', xlabel='Epoch', title='Embedding de ' + str(ev))
        ax2.grid(linestyle='-', linewidth=0.2)
        ax2.xaxis.set_ticks([v for v in range(12)])
        ax2.yaxis.set_ticks(np.linspace(min(h.history['acc'][0], h.history['val_acc'][0]), max(h.history['acc'][-1], h.history['val_acc'][-1], 0.712123), 7))
        ax2.legend(fancybox=True, loc='lower right')
        plt.tight_layout()
        plt.show()
````

````python
modelos_g = []
embedd_sizes_g = np.linspace(8, 256, 7)
for e in np.nditer(embedd_sizes_g):
    modelos_g.append(generar_modelo_g(int(e)))
````

````python
hist_g = []
for mg in modelos_g:
    hist_g.append(entrenar_modelo_g(mg))
````

````python
graficar_g(hist_g)
````

>  A primeras no se consigue una mejora notoria con respecto a la precisión. Existe mejora pero es leve e incrementa según aumenta el tamaño de _embedding_. Lo último tomando como punto de comparación los resultados obtenidos con el conjunto de validación. <br/>
Además se observa otro patrón, el _overfitting_ se produce antes según aumenta el tamaño de embedding, o en otras palabras, el _overfitting_ para modelos con tamaños pequeños de _embedding_ se produce en _epochs_ posteriores a modelos con tamaños de _embedding_ mayor. Por ejemplo para el modelo con _embedding_ de 8 el _overfitting_ se ve en el _epoch_ 6. En el modelo con _embedding_ de 90 el _overfitting_ se produce entre los _epoch_ 2 y 3. Y para el modelo con _embedding_ de 256 el _overfitting_ se produce entre el primer y segundo _epoch_. <br/>
Según aumenta el tamaño de _embedding_ el tiempo de entrenamiento también aumenta.

> i) Sería preferible para este problema de clasificación usar la concatenación como método de combinación, ya que así se tiene todo el espectro de resultados, es decir el vector de _forward states_ y _backward states_. Con ambos vectores la red tiene mayor espectro para determinar las palabras relevantes dentro de todo el contexto apra determinar el tag más adecuado. Al usar otro tipo de métodos para combinar los vectores se pierde información. 

````python
from keras.layers import Bidirectional
model_i = Sequential()
model_i.add(Embedding(input_dim=n_lemmas, output_dim=32, input_length=max_sentence_length))
layer_lstm = CuDNNLSTM(units=100,return_sequences=True)
model_i.add(Bidirectional(layer_lstm,merge_mode='concat'))
model_i.add(Dense(n_labels, activation='softmax'))
model_i.summary()
model_i.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
````

> k) Esta vez se cargarán las palabras en lugar de los _lemmas_, esto porque es deseable que la red tenga la información de cómo se van estructurando las palabras a partir de las letras ya utilizadas, si se usaran los _lemmas_ la red no podría predecir las distintas flexiones que pueda tomar una palabra, por lo que se predeciría texto sin sentido gramatical. <br/>
Primero se pasa todo el _dataset_ a una sola variable que contiene todas las palabras separadas por espacios. Se obtienen los carácteres utilizados en el texto, que posteriormente se utilizarán para predecir. Estos carácteres son codificados con un número identificador para que puedan ser trabajados por la red. <br/>
Se trabajará con el 30% de el _dataset_, oraciones de 40 carácteres y se generarán avanzando de a 5 letras.

````python
dataset = df_ner.loc[:,["word","lemma"]]
text = ' '.join(dataset["word"]).lower() #corpus
null_character = "*"
chars = [null_character]+sorted(list(set(text)))

char_indices = {c: i for i, c in enumerate(chars)}
indices_char = {i: c for i, c in enumerate(chars)}
maxlen = 40
step = 5 
sentences = []
next_chars = []
size = int(len(text)*0.3) #solo un 20% del corpus
for i in range(0, size - maxlen, step):
    sentences.append(text[i: i + maxlen])
    next_chars.append(text[i + maxlen])
````

````python
print('total chars:', len(chars))
print('nb sequences:', len(sentences))
counter = 0
print("\033[4m\033[1m{:41} -> {}\033[0m".format("Oración", "Siguiente carácter"))
for s, n in zip(sentences, next_chars):
    if counter < 3:
        print ("{:41} -> {:4}".format(s, n))
        counter += 1

del counter
````

> En el último _output_ se ve como las oraciones van formandose con saltos de 5 letras.

> l) Se codifican las oraciones con sus números identificadores estimados en [k)](#pregunta_k). Luego se rellena en el inicio las secuencias que no cumplen con la cantidad de carácteres necesarios. <br/>
La red tendrá una capa de embedding de tamaño 16, seguida de una capa GRU de 512 neuronas, con dropout de un 20%. Se termin con una capa _fully connected_ de 72 neuronas con función de activación _softmax_.

````python
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split

dataX = [[char_indices[char] for char in sentence ] for sentence in sentences]
dataY = [char_indices[char] for char in next_chars]

dataX = pad_sequences(maxlen=40, sequences=dataX, padding="pre", value=char_indices[null_character])
dataY = np.asarray([to_categorical(i, num_classes=len(chars)) for i in dataY])
````

````python
from keras.layers import CuDNNGRU,GRU
embedding_vector = 16
model_l = Sequential()
model_l.add(Embedding(input_dim=len(chars), output_dim=embedding_vector, input_length=maxlen))#
model_l.add(CuDNNGRU(units=512,return_sequences=False)) #or GRU
model_l.add(Dropout(0.2))
model_l.add(Dense(len(chars), activation='softmax'))
model_l.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_l.summary()
````

> m) Se añade al entrenamiento la _callback_ ```on_epoch_end(epoch, logs)``` que se encarga de tomar una secuencia del texto con un largo de 40 carácteres. Luego genera 400 letras a partir esta secuencia teniendo como resultado un texto al terminar cada _epoch_. <br/>
La función encargada de generar letras es ```predict_next_char(model, sentence)``` que recibe como parametro el modelo que se está entrenando y la sentencia determinada por ```on_epoch_end(epoch, logs)```. Lo que hace es hacer un _forward pass_ sobre el estado actual del modelo, el resultado de esta operación corresponde a un vector de probabilidades, el cual es usado en la función ```np.random.choice``` a modo de función de probabiliad, para determinar la letra a generar. <br/>
Ya que esta función de generación va ocurriendo según el modelo va mejorando se ve como mejora la cálidad del texto creado por la red.

````python
def predict_next_char(model, sentence):
    """Predict the next character from the current one"""    
    x_pred = [char_indices[char] for char in sentence]
    x_pred = pad_sequences([x_pred], maxlen=maxlen,padding='pre',value=char_indices[null_character])
    preds = model.predict(x_pred, verbose=0)[0]
    next_index = np.random.choice(len(chars), p=preds)
    return indices_char[next_index]
import random,sys
def on_epoch_end(epoch, logs):
    # Function invoked at end of each epoch. Prints generated text.
    print('\n----- Generating text after Epoch: %d' % epoch)
    start_index = random.randint(0, size - maxlen - 1)
    sentence = text[start_index: start_index + maxlen]
    print('----- Generating with seed: "' + sentence + '"')
    sys.stdout.write(sentence)
    for i in range(400):
        next_char = predict_next_char(model_l, sentence)
        sentence = sentence[1:] + next_char #for next character
        sys.stdout.write(next_char)
        sys.stdout.flush()
    return
from keras.callbacks import LambdaCallback
print_callback = LambdaCallback(on_epoch_end=on_epoch_end)
model_l.fit(dataX, dataY, batch_size=256, epochs=25, callbacks=[print_callback])
````
