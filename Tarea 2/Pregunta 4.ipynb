## 4. Transfer Learning

> A lo largo de este _notebook_ se trabajará con el _dataset_ CIFAR-100 que se distribuye en 100 clases, cada una con 600 imágenes. Cada imagen tiene 3 canales de colores con una resolución de 32 píxeles de ancho y 32 píxeles de alto. 

````python
from keras.datasets import cifar100
(x_train, y_train), (x_test, y_test) = cifar100.load_data(label_mode='fine')
````

> Los píxeles tienen formato RGB, por lo que cada uno puede tomar valores de entre 0 y 255. Por ello la normalización de los datos consiste en dividir toda la matriz por 255. <br/>
Para clasificar es necesario transformar las clases a vectores one-hot, esto se logra con la función de _keras_ ```to_categorical```.

````python
from keras.utils import to_categorical
y_train = to_categorical(y_train, num_classes=100)
y_test = to_categorical(y_test, num_classes=100)
x_train = x_train/255.0
x_test = x_test/255.0
````

<a id="a"></a>
> a) Se crea una red convolucional que se usará para clasificar las imágenes. Esta servirá para comparar el desempeño frente a otros modelos que se construirán más adelante. <br/>
La red cuenta con 2 capas de 32 filtros, estos con un tamaño de 3x3 _stride_ de 1, _relu_ como función de activación y modalidad de _same padding_. Seguido por una función de _max pooling_ y 2 capas convolucionales identicas a las anteriores salvo por la cantidad de feature maps, que ene ste caso son 64. Continua con otra función de _max pooling_ y las capas _fully connected_ usadas para clasificar. Se aplica _dropout_ de 50% entre las estas capas.

````python
from keras.models import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,Dropout,Activation
from keras.utils.vis_utils import plot_model
````

````python
model_a = Sequential()
model_a.add(Conv2D(32, (3, 3), padding='same',input_shape=x_train.shape[1:],activation='relu'))
model_a.add(Conv2D(32, (3, 3),padding='same',activation='relu'))
model_a.add(MaxPooling2D(pool_size=(2, 2)))
model_a.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model_a.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model_a.add(MaxPooling2D(pool_size=(2, 2)))
model_a.add(Flatten())
model_a.add(Dense(1024,activation='relu'))
model_a.add(Dropout(0.5))
model_a.add(Dense(100))
model_a.add(Activation('softmax'))
model_a.summary()
````

````python
from keras.optimizers import SGD
import time
````

````python
optimizador_global = SGD(lr=0.01,momentum=0.9)
model_a.compile(loss='categorical_crossentropy', optimizer=optimizador_global, metrics=['accuracy'])
start = time.process_time()
hist_a = model_a.fit(x_train, y_train, batch_size=128,epochs=15,verbose=1, validation_data=(x_test,y_test))
elapsed_a = time.process_time() - start
````

````python
elapsed_a/60
````

>El entrenamiento duró aproximadamente 6 minutos.

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
````

````python
_, (ax1,  ax2) = plt.subplots(1, 2, figsize=(10,3))
ax1.plot(hist_a.history['loss'], label='entrenamiento')
ax1.plot(hist_a.history['val_loss'], label='validación')
ax1.set(ylabel='Loss', xlabel='Epoch', title='Loss')
ax1.grid(linestyle='-', linewidth=0.2)
ax1.legend(fancybox=True, loc='upper right')

ax2.plot(hist_a.history['acc'], label='entrenamiento')
ax2.plot(hist_a.history['val_acc'], label='validación')
ax2.set(ylabel='Accuracy', xlabel='Epoch', title='Accuracy')
ax2.grid(linestyle='-', linewidth=0.2)
ax2.legend(fancybox=True, loc='lower right')
plt.tight_layout()
plt.show()
````

> Existe un claro _overfitting_, ya que la perdida del conjunto de validación disminuye con el paso de los _epochs_ para luego volver a aumentar, mientras la perdida del conjunto de entrenamiento solamente disminuye.  

<a id="b"></a>
> b) A causa de lo observado en [a)](#a) se concluye que hace falta un regularizador para evitar el _overfitting_, por lo tanto se probará la efectividad de usar dropout en distintos lugares de la red. <br/>
Para disminuir el uso de código se crea la función ```plot_b(h, titulo='Regularizadores')```, la cual se encarga de graficar los distintos modelos con los que se experimentará. La otra función auxiliar será ```comparar_b(h, titulo=' ')```, la cual comparará el desempeño de un modelo concreto con el de [a)](#a).

````python
optimizer_b = SGD(lr=0.01,momentum=0.9)
````

````python
def plot_b(h, titulo='Regularizadores'):
    fig, (ax1,  ax2) = plt.subplots(1, 2, figsize=(10,3))
    ax1.plot(h.history['loss'], label='entrenamiento')
    ax1.plot(h.history['val_loss'], label='validación')
    ax1.set(ylabel='Loss', xlabel='Epoch', title=titulo)
    ax1.grid(linestyle='-', linewidth=0.2)
    ax1.legend(fancybox=True, loc='upper right')

    ax2.plot(h.history['acc'], label='entrenamiento')
    ax2.plot(h.history['val_acc'], label='validación')
    ax2.set(ylabel='Accuracy', xlabel='Epoch', title=titulo)
    ax2.grid(linestyle='-', linewidth=0.2)
    ax2.legend(fancybox=True, loc='lower right')
    plt.tight_layout()
    plt.show()

def comparar_b(h, titulo=' '):
    _, (ax1,  ax2) = plt.subplots(1, 2, figsize=(10,3))
    ax1.plot(hist_a.history['val_loss'], label='Original')
    ax1.plot(h.history['val_loss'], label='Regularizadores')
    ax1.set(ylabel='Loss', xlabel='Epoch', title=titulo)
    ax1.grid(linestyle='-', linewidth=0.2)
    ax1.legend(fancybox=True, loc='upper right')

    ax2.plot(hist_a.history['val_acc'], label='Original')
    ax2.plot(h.history['val_acc'], label='Regularizadores')
    ax2.set(ylabel='Accuracy', xlabel='Epoch', title=titulo)
    ax2.grid(linestyle='-', linewidth=0.2)
    ax2.legend(fancybox=True, loc='lower right')
    plt.tight_layout()
    plt.show()
````

<a id="b1"></a>
> La primera configuración a probar es la de colocar el _dropout_ entre las capas convolucionales.

````python
model_b1 = Sequential()
model_b1.add(Conv2D(32, (3, 3), padding='same',input_shape=x_train.shape[1:],activation='relu'))
model_b1.add(Dropout(0.25))
model_b1.add(Conv2D(32, (3, 3),padding='same',activation='relu'))
model_b1.add(MaxPooling2D(pool_size=(2, 2)))
model_b1.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model_b1.add(Dropout(0.25))
model_b1.add(Conv2D(64, (3, 3),padding='same',activation='relu'))
model_b1.add(MaxPooling2D(pool_size=(2, 2)))
model_b1.add(Flatten())
model_b1.add(Dense(1024,activation='relu'))
model_b1.add(Dropout(0.5))
model_b1.add(Dense(100))
model_b1.add(Activation('softmax'))
````

````python
model_b1.compile(loss='categorical_crossentropy', optimizer=optimizer_b, metrics=['accuracy'])
start = time.process_time()
hist_b1 = model_b1.fit(x_train, y_train, batch_size=128,epochs=15,verbose=1, validation_data=(x_test,y_test))
elapsed_b1 = time.process_time() - start
````

````python
plot_b(hist_b1, 'Dropout entre capas convolucionales')
comparar_b(hist_b1, titulo='Comparación con el primer modelo')
````

> El efecto causado por los regularizadores es evidente al comparar la perdida del modelo original con el de ahora. La perdida no vuelve a aumentar en el modelo de los regularizadores como sucedia en [a)](#a). Por otra parte la precisión no disminuye.

> f) Se entrena el mismo modelo de [g)](#g) añadiendo capas de _batch normalization_. Estas se encargan de asegurar que las activaciones de las capas densas tengan una media cercana a cero y una desviación estándar cercana a uno. Los beneficios de hacer este proceso es que al mantener los valores de las activaciones dentro de un rango de valores, el _backpropagation_ reduce las oscilaciones generadas durante el entrenamiento, haciendo la convergencia más rápida. <br/>
Cuando se usan _mini batches_ pequeños se añade un efecto regularizador a la red, ya que se introduce ruido a cada capa durante la normalización. Al tener _mini batches_ grandes la normalización no introduce mayores ruidos. 

````python
model_f = Sequential()
model_f.add(Flatten(input_shape=features_train.shape[1:]))
model_f.add(BatchNormalization())
model_f.add(Dense(1024, activation='relu'))
model_f.add(BatchNormalization())
model_f.add(Dropout(0.5))
model_f.add(Dense(100))
model_f.add(Activation('softmax'))
model_f.compile(optimizer=optimizador_global,loss='categorical_crossentropy', metrics=['accuracy'])
hist_f = model_f.fit(features_train, y_train,epochs=15, batch_size=128,verbose=1,validation_data=(features_test,y_test))
````

````python
_, (ax1,  ax2) = plt.subplots(1, 2, figsize=(10,3))
ax1.plot(hist_f.history['loss'], label='entrenamiento')
ax1.plot(hist_f.history['val_loss'], label='validación')
ax1.set(ylabel='Loss', xlabel='Epoch', title="Loss")
ax1.grid(linestyle='-', linewidth=0.2)
ax1.legend(fancybox=True, loc='upper right')

ax2.plot(hist_f.history['acc'], label='entrenamiento')
ax2.plot(hist_f.history['val_acc'], label='validación')
ax2.set(ylabel='Accuracy', xlabel='Epoch', title='Accuracy')
ax2.grid(linestyle='-', linewidth=0.2)
ax2.legend(fancybox=True, loc='lower right')
plt.tight_layout()
plt.show()

_, (ax1,  ax2) = plt.subplots(1, 2, figsize=(10,3))
ax1.plot(hist_d.history['val_loss'], label='Modelo e)')
ax1.plot(hist_f.history['val_loss'], label='Modelo f)')
ax1.set(ylabel='Loss', xlabel='Epoch', title="Loss")
ax1.grid(linestyle='-', linewidth=0.2)
ax1.legend(fancybox=True, loc='upper right')

ax2.plot(hist_d.history['val_acc'], label='Modelo e)')
ax2.plot(hist_f.history['val_acc'], label='Modelo f)')
ax2.set(ylabel='Accuracy', xlabel='Epoch', title='Accuracy')
ax2.grid(linestyle='-', linewidth=0.2)
ax2.legend(fancybox=True, loc='lower right')
plt.tight_layout()
plt.show()
````

> g) Similar a como se hizo en [d)](#d) sólo se entrenará parte de la red, con la diferencia que ahora se incluirá parte de las capas convolucionales. Cada epoch de el modelo [d)](#d) tarda 5 segundos, y cada epoch de este modelo tarda 25 segundos. El entrenar estas 3 capas extras implica cuadruplicar el tiempo de entrenamiento. Este cambio implica una mejora de rendimiento, el error disminuye y la precisión aumenta para el decimo epoch. Otro detalle es que la curva de entrenamiento y la curva de validación parece divergir, Esto implica que el modelo no generaliza bien.

````python
import numpy as np
input_tensor=Input(shape=x_train.shape[1:])
modelVGG = VGG16(weights='imagenet', include_top=False,input_tensor=input_tensor )
salida_vgg = modelVGG.get_layer('block4_pool').output_shape
model_g = Sequential()
model_g.add(Conv2D(512,(3, 3),input_shape=salida_vgg[1:],activation='relu',padding='same'))
model_g.add(Conv2D(512,(3, 3),activation='relu',padding='same'))
model_g.add(Conv2D(512,(3, 3),activation='relu',padding='same'))
model_g.add(MaxPooling2D((2, 2),strides=(2,2)))    
## dense section
model_g.add(Flatten())
model_g.add(BatchNormalization())
model_g.add(Dense(1024, activation='relu'))
model_g.add(BatchNormalization())
model_g.add(Dropout(0.5))
model_g.add(Dense(100, activation='softmax'))
# delete last 4 layers of VGG16 and transfer the weight to new model
modelVGG.layers.pop() #delete last maxpooling
for i in np.arange(2,-1,-1):
    last = modelVGG.layers.pop() #delete convolutional layers
    model_g.layers[i].set_weights(last.get_weights())
crop_modelVGG = Model(inputs=modelVGG.input, outputs=modelVGG.layers[-1].output)
crop_modelVGG.summary()
````

````python
features_train = crop_modelVGG.predict(x_train)
features_test = crop_modelVGG.predict(x_test)
````
