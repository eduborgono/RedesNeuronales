## 4. Aplicación de una red neuronal en Localización

a) De acorde a la malla asociada a los valores del archivo *IndoorFingerprint.csv*, se clasifican los datos, de acorde al sector al cual pertenecen. Luego, se extraen los indices presentes en el archivo *mask_test.csv* y se utilizan en nuestra data de prueba, ademas de ser eliminados del dataframe principal.

Una vez separados ambos datos, se procede a normalizar los datos.

```python
from keras.models import Sequential
from keras.layers.core import Dense
from keras.optimizers import SGD
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np

df = pd.read_csv("./IndoorFingerprint.csv")
df = df.assign(Cat = (df.X/8).astype(int)+(((df.Y/4).astype(int)-2)*6))
df = df.drop(['C1hA', '0kxZ','tvMX','OlYb','7rk5','F39L','VNSF','tkxI'], axis=1)

mask_test = np.loadtxt('mask_test.csv',dtype="i")
X_test = df.loc[mask_test]
df = df.drop(df.index[mask_test])
df = df.sample(frac=1)
scaler = StandardScaler().fit(df)
X_train = pd.DataFrame(scaler.transform(df),columns=df.columns)
y_train = df.pop('Cat').values.reshape(-1,1)
X_train = X_train.drop(['Cat'],axis=1)

scaler = StandardScaler().fit(X_test)
X_test_scaled = pd.DataFrame(scaler.transform(X_test),columns=X_test.columns)
y_test = X_test.pop('Cat').values.reshape(-1,1)
X_test_scaled = X_test_scaled.drop(['Cat'],axis=1)
```

Una vez realizado todo el proceso de normalización y dejar delimitados que datos han de ser utilizados para pruebas, y que datos serán aplicados para entrenamiento. Se procede a crear la red neuronal, la cual será una variante de la red utilizada en la pregunta 2. esta recibirá como input la categoría y su output seran las coordenadas.

```python
model = Sequential()
model.add(Dense(800, input_dim=y_train.shape[1], kernel_initializer='glorot_uniform',activation='relu'))
model.add(Dense(400, kernel_initializer='glorot_uniform',activation='relu'))
model.add(Dense(200,  kernel_initializer='glorot_uniform',activation='relu'))
model.add(Dense(100, kernel_initializer='glorot_uniform',activation='relu'))
model.add(Dense(50, kernel_initializer='glorot_uniform',activation='relu'))
model.add(Dense(2, kernel_initializer='glorot_uniform',activation='linear'))
sgd = SGD(lr=0.01, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd,loss='mean_squared_error', metrics=["accuracy"])

hist = model.fit(y_train, X_train, epochs=100, verbose=2, validation_data=(y_test, X_test_scaled))
```

```python
listaloss = []
listaval = []
listaacc = []
listavalacc = []
for i in range(100):
    listaloss.append(hist.history['loss'][i])
    listaval.append(hist.history['val_loss'][i])
    listaacc.append(hist.history['acc'][i])
    listavalacc.append(hist.history['val_acc'][i])

plt.plot(range(101)[1:],listaloss,label='loss', linewidth=2)
plt.plot(range(101)[1:],listaval,label='val_loss', linewidth=2)

plt.legend(['loss','val_loss'])
plt.ylabel('MSE')
plt.xlabel('epochs')
plt.grid(1)
plt.show()

plt.plot(range(101)[1:],listaacc,label='acc', linewidth=2)
plt.plot(range(101)[1:],listavalacc,label='val_acc', linewidth=2)

plt.legend(['acc','val_acc'])
plt.ylabel('MSE')
plt.xlabel('epochs')
plt.grid(1)
plt.show()
```

```python

print("Maximos en accuracy:\ntraining accuracy: "+str(max(listaacc))+ "\nvalidation accuracy: "+str(max(listavalacc))+"\nMinimos en loss:\ntraining loss: "+str(min(listaloss))+  "\nvalidation loss: "+str(min(listaval)) )
print("Promedio de accuracy:\ntraining accuracy: "+ str(np.mean(listaacc))+ "\nvalidation accuracy: "+str(np.mean(listavalacc)))
```
